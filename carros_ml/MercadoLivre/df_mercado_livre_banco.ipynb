```python
cnx = mysql.connector.connect(user='root', password='', host='localhost', database='carros_bd_ml')
cursor = cnx.cursor()
```

```python
for index, row in df.iterrows():
    insert_carros_query = "INSERT INTO carros (marca, modelo, ano, cor, tipo_de_combustivel, motor, tipo_de_transmissao, quilometros, portas, price) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
    carros_data = (row['marca'], row['modelo'], row['ano'], row['cor'], row['tipo_de_combustivel'], row['motor'], row['tipo_de_transmissao'], row['quilometros'], row['portas'], row['price'])
    cursor.execute(insert_carros_query, carros_data)
    carros_id = cursor.lastrowid
    
    insert_carros_info_query = "INSERT INTO carros_info (carros_id, com_ipva_pago, aceita_troca, entrega_a_domicilio, unico_dono, com_preco_negociavel, hgienizacao_completa, nao_aceita_troca, com_garantia_mecanica, test_drive_a_domicilio) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
    carros_info_data = (carros_id, row['com_ipva_pago'], row['aceita_troca'], row['entrega_a_domicilio'], row['unico_dono'], row['com_preco_negociavel'], row['higienizacao_completa'], row['nao_aceita_troca'], row['com_garantia_mecanica'], row['test_drive_a_domicilio'])
    cursor.execute(insert_carros_info_query, carros_info_data)

cnx.commit()
```
